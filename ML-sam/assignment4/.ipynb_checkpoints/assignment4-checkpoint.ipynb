```python
lr=0.01;
precision=0.000001;
prev_step_size=1;
max_iter=1000;
iter=0;
gf= lambda x: (x+3)*2;
x=2;
gd=[];
while(iter < max_iter and prev_step_size> precision):
    prev= x;
    x= x- lr*gf(x);
    prev_step_size= abs(x-prev);
    iter+=1;
    print("it:", iter,"x:",x);
    gd.append(x);
```

```python
from matplotlib.pyplot import plot;
```

```python
plot(gd)
```
